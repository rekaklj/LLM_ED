```python
import pandas as pd
```

```python
with open('MNT_train_tags_adjusted_not_meaningful_strings_removed.txt', encoding='utf8') as infile:
    text1 = infile.read()
print(text1[:200])
```

```python
with open('TimeML_train_tags_adjusted_not_meaningful_strings_removed.txt', encoding='utf8') as infile:
    text2 = infile.read()
print(text2[:200])
```

```python
text1 = text1.split('\n\n')
len(text1)
```

```python
def cleaning(sentence):
    sentence = sentence.replace(" n't", "n't")
    sentence = sentence.replace(" 's", "'s")
    sentence = sentence.replace(" 'm", "'m")
    sentence = sentence.replace(" 're", "'re")
    sentence = sentence.replace(" 've", "'ve")
    sentence = sentence.replace(" 'd", "'d")
    sentence = sentence.replace(" .", ".")
    sentence = sentence.replace(" ,", ",")
    sentence = sentence.replace(" !", "!")
    sentence = sentence.replace(" ?", "?")
    sentence = sentence.replace(" ;", ";")
    sentence = sentence.replace(" :", ":")
    sentence = sentence.replace("( ", "(")
    sentence = sentence.replace(" )", ")")        
    return sentence      
```

```python
def sentence_string(annotated_sentence):
    lines_of_annotated_sentence = annotated_sentence.split('\n')
    sentence = []
    for line in lines_of_annotated_sentence:
        splitline = line.split('\t')
        sentence.append(splitline[0])
    sentence_string = ' '.join(sentence)
    return cleaning(sentence_string)
```

```python
def examples(list_of_annotated_sentences):
    examples = []
    for item in list_of_annotated_sentences:
        sentence = sentence_string(item)
        example = 'Sentence:\n' + sentence + '\n\n' + 'Annotated sentence:' + '\n\n' + item
        examples.append(example)
    return examples
```

```python
ex1 = examples(text1)
```

```python
print(ex1[1000])
```

```python
text2 = text2.split('\n\n')
len(text2)
```

```python
ex2 = examples(text2)
```

```python
print(ex2[1000])
```
