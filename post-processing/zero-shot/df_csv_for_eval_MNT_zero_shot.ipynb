```python
ids = []
tokens = []
gold_labels = []
predicted_labels = []
```

```python
line_items = []
for line in text:
    items = line.split()
    line_items.append(items)

print(len(line_items))
print(line_items[:20])
```

```python
print(line_items[0])
```

```python
for line in line_items:
    if len(line) ==5:
        ids.append(line[0])
        tokens.append(line[1])
        gold_labels.append(line[2])
        predicted_labels.append(line[4])
```
