```python
def remove_the_beginning(text):
    if 'Annotated sentence:\n\n' in text:
        removebeginning = text.split("Annotated sentence:\n\n", 1)
        removebeginning[0] = ''
        removebeginning = ''.join(removebeginning)
    elif 'Annotated sentence:\n' in text:
        removebeginning = text.split("Annotated sentence:\n", 1)
        removebeginning[0] = ''
        removebeginning = ''.join(removebeginning)
    elif 'New sentence:\n\n' in text:
        removebeginning = text.split("New sentence:\n\n", 1)
        removebeginning[0] = ''
        removebeginning = ''.join(removebeginning)
    else:
        removebeginning = text
    return removebeginning
```

```python
df['LLM_response_token_tag_only'] = df['LLM_response'].apply(remove_the_beginning)
df.head()
```

```python
def split_negation(sentence):
    sentence = sentence.replace("couldn't\t<O>\n", "could\t<O>\nn't\t<O>\n")
    sentence = sentence.replace("shouldn't\t<O>\n", "should\t<O>\nn't\t<O>\n")
    sentence = sentence.replace("wouldn't\t<O>\n", "would\t<O>\nn't\t<O>\n")
    sentence = sentence.replace("don't\t<O>\n", "do\t<O>\nn't\t<O>\n")
    sentence = sentence.replace("doesn't\t<O>\n", "does\t<O>\nn't\t<O>\n")
    sentence = sentence.replace("isn't\t<O>\n", "is\t<O>\nn't\t<O>\n")
    sentence = sentence.replace("aren't\t<O>\n", "are\t<O>\nn't\t<O>\n")
    sentence = sentence.replace("didn't\t<O>\n", "did\t<O>\nn't\t<O>\n")
    sentence = sentence.replace("wasn't\t<O>\n", "was\t<O>\nn't\t<O>\n")
    sentence = sentence.replace("weren't\t<O>\n", "were\t<O>\nn't\t<O>\n")
    sentence = sentence.replace("hasn't\t<O>\n", "has\t<O>\nn't\t<O>\n")
    sentence = sentence.replace("haven't\t<O>\n", "have\t<O>\nn't\t<O>\n")
    sentence = sentence.replace("hadn't\t<O>\n", "had\t<O>\nn't\t<O>\n")
    if "\n's\t<O>\n" in sentence:
        sentence = sentence.replace("\n's\t<O>\n", "\n's\t<O>\n")
    else:
        sentence = sentence.replace("'s\t<O>\n", "\t<O>\n's\t<O>\n")
    return sentence
```

```python
df['LLM_response_split_negation'] = df['LLM_response_token_tag_only'].apply(split_negation)
df.head()
```

```python
def token_tag(annotated_sentence):
    lines = annotated_sentence.split('\n')
    list_of_token_tag = []
    for line in lines:
        line = line.split()
        if len(line) == 2:
            if 'EVENT' in line[1] or 'O' in line[1]:
                list_of_token_tag.append(line)
                list_of_token_tag
    tuples = []
    for pair in list_of_token_tag:
        tuples.append(tuple(pair))
    return tuple(tuples) # I am using tuples for further processing, see zip()
```

```python
df['list_of_token_tag_pairs'] = df['LLM_response_split_negation'].apply(token_tag)
```

```python
# remove commas and fullstops from tuple lists to make the postprocessing feasible
def remove_commas_fullstops(tuple_of_tuples):
    list_of_tuples = list(tuple_of_tuples)
    if len(tuple_of_tuples) !=0:
        for item in list_of_tuples:
            list_without_commas_and_fullstops = [item for item in list_of_tuples if item[0] != ',' and item[0] != '.']
        return tuple(list_without_commas_and_fullstops)
    else:
        return tuple_of_tuples
    return    
```

```python
def zips(tuples):
    lines = []
    for item in tuples:
        lines.append(f'{id(item)}\t{item[0][0]}\t{item[0][1]}\t{item[1][0]}\t{item[1][1]}\n')
    return ''.join(lines)
```

```python
for idx, row in df.iloc[0:].iterrows():    
    a=row['zippedlines']
    a= zips(a)
    with open('MNT_few_shot_gold_predicted_1706.txt', 'a', encoding='utf8') as outfile:
        outfile.write(f'{a}\n')    
```
